```python
allprojects = requests.get("https://eu-citizen.science/api/projects/")
```

```python
projects = allprojects.json()
projects
```

```python
import re

with open("EUcsImages.txt", 'w') as outfile:
    for url in projects:
        #print(type(url['image1']))
        if url['image1'] is not None:
            correcturl = re.sub('\/media\/media', '/media', url['image1'])
            print(correcturl)
            outfile.write(correcturl + "\n")
```

```python
for p in projects:
    print(p['name'])
    print(p['aim'])
    print(p['description'])
    for k in p['keywords']:
        print(k)
    for t in p['topic']:
        print(t)
    print(p['start_date'])
    print(p['end_date'])
    print(p['url'])
    print(p['latitude'])
    print(p['longitude'])
    print(p['country'])
```

```python
"""Make a file with keywords"""

with open('EUkeywords.txt', 'w') as eukw:
    for p in projects:
        for k in p['keywords']:
            eukw.write(k['keyword'] + "\n")
```
