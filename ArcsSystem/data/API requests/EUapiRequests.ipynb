```python
import requests
import json
from xml.sax.saxutils import escape
import csv
def replaceslash(keyword):
    if "/" in keyword:
        return keyword.replace("/", "&amp;")
    else:
        return(keyword)
```

```python
#parameters = {"/projects/"}
response = requests.get("https://eu-citizen.science/api/projects/")
```

```python
def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)
```

```python
for project in response.json():
    if project['country'] == "SE":
        print(project)
```

```python
existing_list = []
with open('../publications/top500keywordsSVENtranslated.csv', 'r') as thefile:
        data = csv.DictReader(thefile, delimiter=',', )
        for d in data:
            #print(d['en'])
            existing_list.append(d['en'])
print(existing_list)
```

```python
keywordlist = []

for project in response.json():
    for k in project['keywords']:
        if k['keyword'] not in keywordlist:
            keywordlist.append(k['keyword'])
            
for project in response.json():
    for k in project['topic']:
        #print(k['topic'])
        if k['topic'] not in keywordlist:
            keywordlist.append(k['topic'])        
            
            
for kw in keywordlist:
    correctkw = replaceslash(escape(kw).lower())
    if correctkw in existing_list:
        continue
    else:
        print(correctkw)
```
