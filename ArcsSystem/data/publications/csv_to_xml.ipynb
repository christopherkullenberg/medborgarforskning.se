```python
import re
import sys
import json
import csv
from xml.sax.saxutils import escape
import requests



def replaceslash(keyword):
    if "/" in keyword:
        return keyword.replace("/", "&amp;")
    else:
        return(keyword)
    
def localimage(url):
    filename = re.sub('https\:\/\/eu\-citizen.science\/media\/media\/images\/',
                        '', url)
    #removeampersand = re.sub('\&', '&amp;', filename)
    localurl = 'EUimg/' + filename
    return localurl

def remoteimage(value):
    remoteurl = re.sub('\/media\/media\/',
                        '/media/', value)
    return remoteurl


       

def parsecsv():
    with open('ARCSLiterature.csv', 'r') as thefile:
        data = csv.DictReader(thefile, delimiter=',', quotechar='"')
        datadict = {}
        for row in data:
            if escape(row['Title']) not in datadict:
                datadict[row['Title']] = [row['Author'],
                                                 row['Publication Year'],
                                                 row['Publication Title'],
                                                 row['Issue'],
                                                 row['Volume'],
                                                 row['DOI'],
                                                 row['Abstract Note'],
                                                 row['Manual Tags']
                                          ]

            else:
                continue
        return datadict

    

    
    

#valueslist = []
```

```python
response = requests.get("https://eu-citizen.science/api/projects/")
print(response.text)


def parseEUprojects():
    datadict = {}
    for project in response.json():
        if escape(project['name']) not in datadict:
            datadict[project['name']] = [project['aim'],
                                 project['keywords'],
                                 project['topic'],
                                 project['url'],
                                 project['image1'],
                                 project['country'],
                                 project['dateCreated'],
                                          ]
        else:
                continue
                
    return datadict

        
        
      


        
parseEUprojects()
```

```python
wikiQdict = {}

def makeQdict():
    with open('keyword_wikiQs.csv', 'r') as qfile:
        data = csv.DictReader(qfile, delimiter=',', quotechar='"')
        for d in data:
            wikiQdict[d["keyword"]] = d["Q"][1:]

makeQdict()
for k, v in wikiQdict.items():
    print(k,v)

```

```python
summarydict = {}

def makesummarydict():
    with open("wikipediasummaryofkeywords.csv", "r") as sumfile:
        data = csv.DictReader(sumfile, delimiter=',', quotechar='"')
        for d in data:
            summarydict[d["keyword"]] = d["summary"]

makesummarydict()

summarydict["american pika"]
    
```

```python
keyworddict = {}

counter = 0
print(counter)

for k, v in parsecsv().items():
    keywordlist = v[7].split("; ")
    for x in keywordlist:

        loweredkeyword = replaceslash(escape(x.lower()))
        if len(loweredkeyword) > 1:
            if loweredkeyword not in keyworddict:
                print(loweredkeyword)
                keyworddict[loweredkeyword] = counter
                counter += 1
                print(counter)
                
for k, v in parseEUprojects().items():
    for key in v[1]:
        keywordlist = key['keyword']
        #print("Keywords: ", keywordlist)
        loweredpkeyword = replaceslash(escape(keywordlist.lower()))
        #print(loweredkeyword)
        if len(loweredpkeyword) > 1:
            if loweredpkeyword not in keyworddict:
                #print(loweredkeyword)
                keyworddict[loweredpkeyword] = counter
                print(loweredpkeyword)
                counter += 1
                print(counter)

    for key in v[2]:
        topiclist = key['topic']
        loweredtopic = replaceslash(escape(topiclist.lower()))
        #print(loweredtopic)
        if len(loweredtopic) > 1:
            if loweredtopic not in keyworddict:
                print(loweredtopic)
                keyworddict[loweredtopic] = counter
                counter += 1
 

                print(counter)

#print(sorted(keyworddict))
```

```python
# open up dropped papers and define keyword lookup function

pkcounter = 1

droppedpapers = open('droppedpapers.txt', 'w')

def kwlookup(kwlist):   
    kwlookuplist = []

    for x in keywordlist:
        loweredkeyword = x
        #print(loweredkeyword)

        if loweredkeyword in keyworddict:
            #print("\t", loweredkeyword)
            #print(keyworddict[loweredkeyword])
            #print(loweredkeyword)
            kwlookuplist.append(keyworddict[loweredkeyword])
        else:
            continue
    return(kwlookuplist)
```
