```python
import re
import sys
import json
import csv
from xml.sax.saxutils import escape
import requests



def replaceslash(keyword):
    if "/" in keyword:
        return keyword.replace("/", "&amp;")
    else:
        return(keyword)
    
def localimage(url):
    filename = re.sub('https\:\/\/eu\-citizen.science\/media\/media\/images\/',
                        '', url)
    #removeampersand = re.sub('\&', '&amp;', filename)
    localurl = 'EUimg/' + filename
    return localurl

def remoteimage(value):
    remoteurl = re.sub('\/media\/media\/',
                        '/media/', value)
    return remoteurl


       

def parsecsv():
    with open('ARCSLiterature.csv', 'r') as thefile:
        data = csv.DictReader(thefile, delimiter=',', quotechar='"')
        datadict = {}
        for row in data:
            if escape(row['Title']) not in datadict:
                datadict[row['Title']] = [row['Author'],
                                                 row['Publication Year'],
                                                 row['Publication Title'],
                                                 row['Issue'],
                                                 row['Volume'],
                                                 row['DOI'],
                                                 row['Abstract Note'],
                                                 row['Manual Tags'],
                                                
                                          ]

            else:
                continue
        return datadict

    

    
    

#valueslist = []
```

```python
response = requests.get("https://eu-citizen.science/api/projects/")
print(response.text)


def parseEUprojects():
    datadict = {}
    for project in response.json():
        if escape(project['name']) not in datadict:
            datadict[project['name']] = [project['aim'],
                                 project['keywords'],
                                 project['topic'],
                                 project['url'],
                                 project['image1'],
                                 project['country'],
                                 project['dateCreated'],
                                 project['start_date'],
                                 project['end_date'], 
                                 project['latitude'],
                                 project['longitude'],
                                          ]
        else:
                continue
                
    return datadict

        
        
      


        
parseEUprojects()
```

```python
wikiQdict = {}

def makeQdict():
    with open('keyword_wikiQs.csv', 'r') as qfile:
        data = csv.DictReader(qfile, delimiter=',', quotechar='"')
        for d in data:
            wikiQdict[d["keyword"]] = d["Q"][1:]

makeQdict()
for k, v in wikiQdict.items():
    print(k,v)

```

```python
# open up dropped papers and define keyword lookup function

pkcounter = 1

droppedpapers = open('droppedpapers.txt', 'w')

def kwlookup(kwlist):   
    kwlookuplist = []
    #print(kwlist)
    for k, v in translation_line_dict.items():
        for x in kwlist:
            loweredkeyword = x
            #print(loweredkeyword)

            if loweredkeyword in keyworddict:
                if keyworddict[loweredkeyword] == v[1]:
                    #print("\t", loweredkeyword)
                    #print(keyworddict[loweredkeyword])
                    #print(loweredkeyword)
                    kwlookuplist.append(k)
            else:
                continue
    return(kwlookuplist)
```
